Como hemos podido observar "age" viene como [[60.0, 'year']] y "gender" como M/F.  

El siguiente paso ser√° normalizar "age" y "gender":

- age_col_norm (int) 
     
    "[[60.0, 'year']]"  -> 60
    "60" / "60.0"       -> 60
    Devuelve int o None si no encuentra un valor razonable (0<edad<130).

- sex_col_norm (male/female)

    Mapea variantes a {'male','female'}.
    Acepta: 'M','F','male','female','man','woman','hombre','mujer', etc.

El obejtivo es tener referecnias limpias a partir de los que ya viene estructurado en el CSV, para poder comparar contra lo que se extraiga del texto libre de la oclumna 'patient'.

```python
import re

def parse_age_col(val:str):
    if not val or not isinstance(val, str):
        return None
    m = re.search(r'([0-9]+(?:\.[0-9]+)?)',val)
    if not m:
        return None
    try:
        num = float(m.group(1))
        if 0 < num < 130:
            return int(round(num))
    except:
        return None
    return None

def normalize_gender_col(val:str):
    if not val:
        return None
    v = val.strip().lower()
    if v in {"m", "male", "man", "hombre"}:
        return "male"
    if v in {"f", "female", "woman", "mujer"}:
        return "female"
    return None

#Creamos las columnas normalizadas:

df["age_col_norm"] = pd.to_numeric(df["age_col_norm"], errors="coerce").astype("Int64")
df["sex_col_norm"] = df.get("gender", "").apply(normalize_gender_col)

#Vista
try:
    display(df[["age", "age_col_norm", "gender", "sex_col_norm"]].head(5))
except NameError:
    print(df[["age","age_col_norm","gender","sex_col_norm"]].head(5))
```

El siguiente paso ser√° extraer edad y sexo desde la columna "patient" y vamos a compararlas con las columnas normalizadas

```python
# %% [markdown]
# CELDA 6 ‚Äî ANTES
# Objetivo: Definir las regex y la funci√≥n `extract_age_sex_from_text(note_text)`.
# Esta celda no depende de N_SAMPLE y la usaremos en TODAS las filas.

# %%
import re

# Patrones b√°sicos que aparecen en tus textos:
# - "60-year-old", "39-year-old"
# - "sex: male", "gender: female", etc. (incluye variantes en espa√±ol)
age_re_1 = re.compile(r'(?i)\b([0-9]{1,3})-year-old\b')
age_re_2 = re.compile(r'(?i)\bage\s*[:=]?\s*([0-9]{1,3})\b')
sex_re   = re.compile(r'(?i)\b(sex|gender)\s*[:=]?\s*(male|female|man|woman|hombre|mujer|m|f)\b')

def extract_age_sex_from_text(note_text: str):
    """
    Devuelve dict con:
      - age_from_text (int o None)
      - sex_from_text (string original o None)
      - sex_from_text_norm ('male'/'female' o None)
    Usa tu normalize_gender_col() para normalizar el sexo.
    """
    t = (note_text or "").replace("\u2009"," ").replace("\u00a0"," ")
    out = {}

    # Edad (primero "X-year-old", si no, "age: X")
    m = age_re_1.search(t) or age_re_2.search(t)
    if m:
        try:
            out["age_from_text"] = int(m.group(1))
        except:
            out["age_from_text"] = None
    else:
        out["age_from_text"] = None

    # Sexo (normalizamos a male/female con tu funci√≥n)
    m = sex_re.search(t)
    if m:
        out["sex_from_text"] = m.group(2).lower()
        out["sex_from_text_norm"] = normalize_gender_col(out["sex_from_text"])
    else:
        out["sex_from_text"] = None
        out["sex_from_text_norm"] = None

    return out

# Prueba r√°pida opcional en 2 filas (para ver formato del dict):
try:
    for txt in df["patient"].head(2):
        print(extract_age_sex_from_text(txt))
except Exception as e:
    print("Prueba r√°pida: omitiendo por error:", e)

# %% [markdown]
# CELDA 6 ‚Äî DESPU√âS
# Ya existe `extract_age_sex_from_text`. Ahora puedes ejecutar tu CELDA 7 para procesar todas las filas.
```

```python
# %% [markdown]
# CELDA 7A ‚Äî ANTES
# Objetivo: Mejorar la extracci√≥n de SEXO desde el texto `patient`.
# - Capturamos: "60-year-old male", "A 39-year-old man", "female patient", etc.
# - Evitamos "female-to-male" con lookarounds que descartan guiones.

# %%
import re

# 1) Edad + sexo juntos: "60-year-old male", "57 year old woman", con opcional coma/descriptores
age_sex_re = re.compile(
    r"(?i)\b([0-9]{1,3})\s*[- ]?year[- ]?old\b[ ,;:\-]*\s*(male|female|man|woman)\b"
)

# 2) "male patient" / "female patient"
patient_sex_re = re.compile(r"(?i)\b(male|female)\s+patient\b")

# 3) Palabra suelta male/female/man/woman, evitando cosas como "female-to-male"
general_sex_re = re.compile(r"(?i)(?<!-)\b(male|female|man|woman)\b(?!-)")

def extract_sex_from_text_v2(note_text: str):
    """
    Devuelve:
      - sex_from_text_v2 (string original encontrado)
      - sex_from_text_norm_v2 ('male'|'female'), mapeando 'man'->'male' y 'woman'->'female'
    En orden de fiabilidad: age_sex_re > patient_sex_re > general_sex_re
    """
    t = (note_text or "").replace("\u2009"," ").replace("\u00a0"," ")
    for rx in (age_sex_re, patient_sex_re, general_sex_re):
        m = rx.search(t)
        if m:
            token = m.group(1).lower() if rx is not age_sex_re else m.group(2).lower()
            if token in {"man", "male"}:
                return token, "male"
            if token in {"woman", "female"}:
                return token, "female"
    return None, None

# Prueba r√°pida en 2 textos (opcional)
try:
    for txt in df["patient"].head(2):
        print(extract_sex_from_text_v2(txt))
except Exception as e:
    print("Prueba r√°pida: omito por error:", e)

# %% [markdown]
# CELDA 7A ‚Äî DESPU√âS
# Ya tenemos un extractor mejorado para el SEXO.
# En la siguiente celda lo aplicamos a TODO el dataset y medimos `sex_match`.
```

```python
# %% [markdown]
# üîó CELDA 6-ter ‚Äî ANTES
# Objetivo: Combinar la extracci√≥n de edad (la tuya) con el sexo v2 en una sola funci√≥n.

# %%
def extract_age_sex_from_text_v2(note_text: str):
    t = (note_text or "").replace("\u2009"," ").replace("\u00a0"," ")
    out = {}

    # Edad (igual que tu celda)
    m = age_re_1.search(t) or age_re_2.search(t)
    out["age_from_text"] = int(m.group(1)) if m else None

    # Sexo (usa v2 + tu normalizador)
    tok, tok_norm = extract_sex_from_text_v2(t)
    out["sex_from_text"] = tok
    out["sex_from_text_norm"] = tok_norm

    return out

# Prueba r√°pida:
try:
    for txt in df["patient"].head(2):
        print(extract_age_sex_from_text_v2(txt))
except Exception as e:
    print("Prueba r√°pida: omito por error:", e)

# %% [markdown]
# CELDA 6-ter ‚Äî DESPU√âS
# Ya tienes una √∫nica funci√≥n con edad + sexo (v2).
```

El siguiente paso ser√° ver en qu√© porcentaje notas o filas conseguimos extraer el sexo y la edad desde el texto. De los casos en los que se extrae sexo y edad del texto, ¬øCu√°ntos coinciden con las columnas sexo y g√©nero respectivamente?

```python
# CELDA 7B ‚Äî ANTES
# Objetivo: Evaluar **edad y sexo** extra√≠dos desde el texto `patient` frente a las columnas normalizadas del CSV.
# M√©tricas que calculamos:
#   - Cobertura de extracci√≥n desde texto (qu√© % de filas lo traen).
#   - Exactitud (match) ignorando NaN: proporci√≥n de coincidencias cuando hay dato en ambos lados.
# Notas:
#   - Para SEXO: si existe extract_sex_from_text_v3() (con pronombres), lo usaremos; si no, usamos v2.
#   - Para EDAD: usamos los patrones age_re_1 / age_re_2 ya definidos arriba.

# %%
import pandas as pd
import numpy as np
import time
import re

# --- 1) Asegurar funciones/patrones para EDAD ---
# Si por lo que sea no existen en el entorno, los definimos aqu√≠ de forma segura.
if 'age_re_1' not in globals():
    age_re_1 = re.compile(r'(?i)\b([0-9]{1,3})-year-old\b')
if 'age_re_2' not in globals():
    age_re_2 = re.compile(r'(?i)\bage\s*[:=]?\s*([0-9]{1,3})\b')

def extract_age_from_text(note_text: str):
    """Devuelve edad (int) si encuentra 'X-year-old' o 'age: X'; en caso contrario None."""
    t = (note_text or "").replace("\u2009"," ").replace("\u00a0"," ")
    m = age_re_1.search(t) or age_re_2.search(t)
    if not m:
        return None
    try:
        return int(m.group(1))
    except:
        return None

# --- 2) Envoltorio para SEXO: usa v3 si existe; si no, v2 ---
def get_sex_from_text(note_text: str):
    """
    Devuelve tupla estandarizada:
      (sex_token, sex_norm, method)
      - sex_token: texto encontrado ('male','man', 'pronoun-male', etc.) o None
      - sex_norm : 'male'/'female' o None
      - method   : 'explicit' (v2) o 'pronoun_fallback'/'explicit' (v3)
    """
    t = (note_text or "").replace("\u2009"," ").replace("\u00a0"," ")
    # v3 disponible (pronombres)
    if 'extract_sex_from_text_v3' in globals():
        tok, tok_norm, method = extract_sex_from_text_v3(t)
        return tok, tok_norm, method
    # si no hay v3, usamos v2 (solo expl√≠cito)
    elif 'extract_sex_from_text_v2' in globals():
        tok, tok_norm = extract_sex_from_text_v2(t)
        method = 'explicit' if tok_norm else None
        return tok, tok_norm, method
    else:
        # seguridad por si faltara todo
        return None, None, None

# --- 3) Bucle por TODO el dataset ---
t0 = time.time()
rows = []

for r in df.itertuples(index=False):
    pid          = getattr(r, "patient_id", "")
    text         = getattr(r, "patient", "")
    age_col_norm = getattr(r, "age_col_norm", None)
    sex_col_norm = getattr(r, "sex_col_norm", None)

    # Extraer EDAD desde texto
    age_txt = extract_age_from_text(text)

    # Extraer SEXO desde texto (v3 si existe, si no v2)
    sex_tok, sex_norm, sex_method = get_sex_from_text(text)

    rows.append({
        "patient_id": pid,
        # Ground truth (CSV normalizado)
        "age_col_norm": age_col_norm,
        "sex_col_norm": sex_col_norm,
        # Texto
        "age_from_text": age_txt,
        "sex_from_text": sex_tok,
        "sex_from_text_norm": sex_norm,
        "sex_method": sex_method,
    })

cmp_df = pd.DataFrame(rows)

# --- 4) Tipos num√©ricos para EDAD (comparaci√≥n limpia) ---
cmp_df["age_col_norm_int"]  = pd.to_numeric(cmp_df["age_col_norm"], errors="coerce").astype("Int64")
cmp_df["age_from_text_int"] = pd.to_numeric(cmp_df["age_from_text"], errors="coerce").astype("Int64")

# --- 5) M√©tricas de COBERTURA desde texto ---
age_cov_text = cmp_df["age_from_text_int"].notna().mean()                 # % de filas con edad extra√≠da del texto
sex_cov_text = cmp_df["sex_from_text_norm"].notna().mean()                # % de filas con sexo extra√≠do del texto

# --- 6) M√©tricas de EXACTITUD (match) ignorando NaN ---
cmp_df["age_match"] = np.where(
    cmp_df["age_from_text_int"].notna() & cmp_df["age_col_norm_int"].notna(),
    cmp_df["age_from_text_int"] == cmp_df["age_col_norm_int"],
    np.nan
)
cmp_df["sex_match"] = np.where(
    cmp_df["sex_from_text_norm"].notna() & cmp_df["sex_col_norm"].notna(),
    cmp_df["sex_from_text_norm"] == cmp_df["sex_col_norm"],
    np.nan
)

age_acc = cmp_df["age_match"].mean(skipna=True)  # exactitud condicional de edad
sex_acc = cmp_df["sex_match"].mean(skipna=True)  # exactitud condicional de sexo

t1 = time.time()

# --- 7) Reporte resumido ---
print(f"Filas procesadas: {len(cmp_df)}   Tiempo: {t1 - t0:.1f}s")
print(f"COBERTURA desde texto -> edad: {age_cov_text:.2%} | sexo: {sex_cov_text:.2%}")
print(f"EXACTITUD (match, ignora NaN) -> edad: {age_acc:.2%} | sexo: {sex_acc:.2%}")

# Muestra algunas filas para inspecci√≥n
cols_show = [
    "patient_id",
    "age_col_norm_int","age_from_text_int","age_match",
    "sex_col_norm","sex_from_text","sex_from_text_norm","sex_method","sex_match"
]
try:
    display(cmp_df[cols_show].head(10))
except NameError:
    print(cmp_df[cols_show].head(10))
```


En el siguiente paso vamos a convertir altura a metros y a normalizar los decimales (70,5 -> 70.5)


```python
# CELDA 8A ‚Äî ANTES
# Objetivo: utilidades de conversi√≥n de altura a metros y normalizaci√≥n de decimales.

# %%
import re

def _norm_decimals(s: str) -> str:
    """Convierte '70,5' -> '70.5' solo cuando la coma separa d√≠gitos."""
    return re.sub(r'(?<=\d),(?=\d)', '.', s or "")

def height_to_m(value, unit):
    """Convierte altura (m/cm/mm/in/ft) a metros. Devuelve float o None."""
    try:
        v = float(value)
    except:
        return None
    u = (unit or "").lower().strip()
    if u in {"m","meter","meters"}:
        return v
    if u in {"cm","centimeter","centimeters"}:
        return v / 100.0
    if u in {"mm","millimeter","millimeters"}:
        return v / 1000.0
    if u in {"in","inch","inches"}:
        return v * 0.0254
    if u in {"ft","foot","feet"}:
        return v * 0.3048
    return None

def feet_inches_to_m(feet, inches=0):
    """Convierte pies+pulgadas a metros (e.g., 5'11\" -> m)."""
    try:
        f = float(feet); i = float(inches or 0)
    except:
        return None
    return f * 0.3048 + i * 0.0254

# CELDA 8A ‚Äî DESPU√âS
# Listas las utilidades de altura. Seguimos con la extracci√≥n desde `patient`.
```

Extracci√≥n de Altura desde patient.

Regex para muchos formatos:

height/ht: 172 cm, ht 1.72 m

170 cm tall, 1.70 m tall

5'11" o 5 ft 11 in (o solo pies)

stature/body height ... 170 cm

Normalizaremos a metros y guardamos source para trazabilidad



```python
# CELDA 8B ‚Äî ANTES
# Objetivo: Extraer altura con m√∫ltiples patrones y normalizar a metros.

# %%
# Patrones "m√©tricos" etiquetados
height_metric_re = re.compile(
    r'(?i)\b(?:height|ht)\b\s*[:=]?\s*([0-9]+(?:[.,][0-9]+)?)\s*(cm|centimeters?|m|meters?|mm|millimeters?)\b'
)

# "170 cm tall" / "1.70 m tall"
height_tall_re = re.compile(
    r'(?i)\b([0-9]+(?:[.,][0-9]+)?)\s*(cm|centimeters?|m|meters?)\s+tall\b'
)

# Contexto "stature" / "body height"
height_context_re = re.compile(
    r'(?i)\b(?:stature|body\s*height)\b[^0-9]{0,20}([0-9]+(?:[.,][0-9]+)?)\s*(cm|centimeters?|m|meters?|mm|millimeters?)\b'
)

# Pies/pulgadas: 5'11"  (permite 4‚Äì7 pies, 0‚Äì11 pulgadas)
height_ftin_quote_re = re.compile(
    r'(?i)\b([4-7])\'\s*([0-9]{1,2})(?:"|\'\')?'
)

# Pies/pulgadas con palabras: 5 ft 11 in
height_ftin_words_re = re.compile(
    r'(?i)\b([4-7])\s*(?:ft|feet|foot)\s*([0-9]{1,2})\s*(?:in|inch|inches)\b'
)

# Solo pies: 6 ft
height_ft_only_re = re.compile(
    r'(?i)\b([4-7])\s*(?:ft|feet|foot)\b'
)

def extract_height_from_text(note_text: str):
    """
    Devuelve dict con:
      - height_m (float)
      - height_raw (traza: value/unit)
      - height_source (patr√≥n que lo detect√≥)
    """
    t = _norm_decimals((note_text or "")).replace("\u2009"," ").replace("\u00a0"," ")

    # M√©trico etiquetado
    m = height_metric_re.search(t)
    if m:
        val, unit = m.group(1), m.group(2)
        hm = height_to_m(val, unit)
        if hm is not None:
            return {"height_m": hm, "height_raw": {"value": float(val.replace(',', '.')), "unit": unit}, "height_source": "metric_tag"}

    # "cm/m tall"
    m = height_tall_re.search(t)
    if m:
        val, unit = m.group(1), m.group(2)
        hm = height_to_m(val, unit)
        if hm is not None:
            return {"height_m": hm, "height_raw": {"value": float(val.replace(',', '.')), "unit": unit}, "height_source": "metric_tall"}

    # Contexto "stature/body height"
    m = height_context_re.search(t)
    if m:
        val, unit = m.group(1), m.group(2)
        hm = height_to_m(val, unit)
        if hm is not None:
            return {"height_m": hm, "height_raw": {"value": float(val.replace(',', '.')), "unit": unit}, "height_source": "metric_context"}

    # Pies + pulgadas (con comillas)
    m = height_ftin_quote_re.search(t)
    if m:
        hm = feet_inches_to_m(m.group(1), m.group(2))
        if hm is not None:
            return {"height_m": hm, "height_raw": {"value": f"{m.group(1)}'{m.group(2)}\"", "unit": "ft_in"}, "height_source": "ft_in_quote"}

    # Pies + pulgadas (con palabras)
    m = height_ftin_words_re.search(t)
    if m:
        hm = feet_inches_to_m(m.group(1), m.group(2))
        if hm is not None:
            return {"height_m": hm, "height_raw": {"value": f"{m.group(1)} ft {m.group(2)} in", "unit": "ft_in"}, "height_source": "ft_in_words"}

    # Solo pies
    m = height_ft_only_re.search(t)
    if m:
        hm = feet_inches_to_m(m.group(1), 0)
        if hm is not None:
            return {"height_m": hm, "height_raw": {"value": f"{m.group(1)} ft", "unit": "ft"}, "height_source": "ft_only"}

    return {}

# Prueba r√°pida (3 textos)
try:
    for txt in df["patient"].head(3):
        print(extract_height_from_text(txt))
except Exception as e:
    print("Prueba r√°pida omitida:", e)

# CELDA 8B# El extractor devuelve metros + traza + source. Si no hay altura en el texto, devuelve {}.
```

```python
#ALTURA: pistas vs extracci√≥n
# =========================
import re, pandas as pd, numpy as np

# --- RUTAS (¬°ajusta estas dos a tu entorno!)
DATA_CSV = r"C:/Users/hered/Desktop/TFM/TFM/data/PMC-Patients.csv"
HEIGHT_BASELINE_CSV = r"C:/Users/hered/Desktop/TFM/TFM/baseline_height_all.csv"   # el CSV que se gener√≥ con height_m, height_source, etc.

# --- Cargar dataset principal
df = pd.read_csv(DATA_CSV, dtype=str, encoding="utf-8", na_filter=False)
if "patient_id" not in df.columns:
    # Si no hay patient_id, crea uno incremental
    df["patient_id"] = np.arange(len(df)).astype(str)

# --- Cargar baseline de altura (el que generaste)
hb = pd.read_csv(HEIGHT_BASELINE_CSV, dtype=str, encoding="utf-8", na_filter=False)
# Normalizar tipos
hb["height_m"] = pd.to_numeric(hb.get("height_m", np.nan), errors="coerce")
hb["has_extracted_height"] = hb["height_m"].notna()

# --- Unir por patient_id
merged = df.merge(hb[["patient_id","height_m","height_raw","height_source","has_extracted_height"]],
                  on="patient_id", how="left")

# --- Normalizar texto
def norm_text(s: str) -> str:
    s = (s or "")
    s = s.replace("\u2009"," ").replace("\u00a0"," ")
    return s

merged["patient_norm"] = merged["patient"].apply(norm_text)

# Definir PISTAS de altura (cues)
# Nota: mantenemos expresiones ‚Äúlaxas‚Äù para estimar techo de cobertura
cues_patterns = {
    "label_height": re.compile(r"(?i)\b(height|ht\.?|altura|estatura|talla)\b"),
    "stature":      re.compile(r"(?i)\bstature\b"),
    "standing_height": re.compile(r"(?i)\bstanding\s+height\b"),
    "body_height":  re.compile(r"(?i)\bbody\s*height\b"),
    "body_length":  re.compile(r"(?i)\b(?:body\s*)?length\b"),
    # valores m√©tricos plausibles cerca de cm/m
    "cm_value":     re.compile(r"(?i)\b(?:[3-9]\d|1\d{2}|2[0-4]\d|250)\s*cm\b"),  # 30-250 cm
    "m_value":      re.compile(r"(?i)\b(?:0\.\d{1,2}|1(?:\.\d{1,2})?|2(?:\.\d{1,2})?)\s*m\b"),  # 0.x-2.x m
    # pies/pulgadas
    "ft_words":     re.compile(r"(?i)\b[4-7]\s*(?:ft|feet|foot)\b"),
    "ft_quotes":    re.compile(r"(?i)\b[4-7]\s*['\u2032]"),
    "in_words":     re.compile(r"(?i)\b[0-9]{1,2}\s*(?:in|inch|inches)\b"),
    # expresiones ‚Äútall‚Äù
    "tall_metric":  re.compile(r"(?i)\b[0-9]+(?:[.,][0-9]+)?\s*(?:cm|centimeters?|m|meters?)\s+tall\b"),
    "tall_inches":  re.compile(r"(?i)\b[0-9]{1,2}\s*(?:in|inch|inches)\s+tall\b")
}

def find_cues(text: str):
    text = text or ""
    hits = []
    for name, rx in cues_patterns.items():
        if rx.search(text):
            hits.append(name)
    return hits

def first_snippets(text: str, max_hits=2, window=60):
    """Devuelve hasta 2 fragmentos alrededor de los primeros matches de cualquier cue."""
    text = text or ""
    spans = []
    for rx in cues_patterns.values():
        for m in rx.finditer(text):
            spans.append((m.start(), m.end()))
    spans.sort(key=lambda x: x[0])
    out = []
    for s,e in spans[:max_hits]:
        start = max(0, s - window)
        end   = min(len(text), e + window)
        out.append(text[start:end].replace("\n"," "))
    return out

merged["height_cues"] = merged["patient_norm"].apply(find_cues)
merged["has_height_cues"] = merged["height_cues"].apply(lambda L: len(L) > 0)
merged["height_cues_snippets"] = merged["patient_norm"].apply(first_snippets)

# --- M√©tricas globales
n = len(merged)
cue_rate = merged["has_height_cues"].mean()
extract_rate = merged["has_extracted_height"].fillna(False).mean()
both_rate = (merged["has_height_cues"] & merged["has_extracted_height"].fillna(False)).mean()

print(f"Filas: {n}")
print(f"- % con PISTAS de altura (cues): {cue_rate:.2%}")
print(f"- % con ALTURA extra√≠da (baseline): {extract_rate:.2%}")
print(f"- % con ambas (pistas y extracci√≥n): {both_rate:.2%}")

# --- Tabla de frecuencias de cada cue
from collections import Counter
all_cues = Counter(c for L in merged["height_cues"] for c in L)
cue_freq = pd.DataFrame(sorted(all_cues.items(), key=lambda x: x[1], reverse=True),
                        columns=["cue","count"])
cue_freq["pct"] = (cue_freq["count"] / n * 100).round(2)
print("\nTop cues:")
print(cue_freq.head(20))

# --- Ejemplos a revisar: hay pistas pero NO se extrajo altura
to_review = merged[(merged["has_height_cues"]) & (~merged["has_extracted_height"].fillna(False))].copy()
cols_view = ["patient_id","height_cues","height_cues_snippets","patient_norm"]
print(f"\nEjemplos con PISTAS pero SIN extracci√≥n (muestra 10 de {len(to_review)}):")
display(to_review[cols_view].head(10))

# --- (Opcional) Ejemplos con extracci√≥n (para ver qu√© patrones s√≠ capturamos)
ok_examples = merged[merged["has_extracted_height"].fillna(False)].copy()
ok_cols = ["patient_id","height_m","height_raw","height_source","height_cues","height_cues_snippets","patient_norm"]
print(f"\nEjemplos con extracci√≥n OK (muestra 10 de {len(ok_examples)}):")
display(ok_examples[ok_cols].head(10))

# --- Guardar informe completo para depurar offline
out_cols = ["patient_id","has_height_cues","height_cues","height_cues_snippets",
            "has_extracted_height","height_m","height_raw","height_source"]
report = merged[out_cols].copy()
report.to_csv("height_cues_vs_extraction.csv", index=False, encoding="utf-8")
print("\nGuardado informe: height_cues_vs_extraction.csv")

# --- Resumen r√°pido por tipos de cue (¬øen cu√°les fallamos m√°s?)
fail = to_review.explode("height_cues")
fail_counts = (fail.groupby("height_cues")["patient_id"]
               .count()
               .sort_values(ascending=False)
               .rename("fails"))
summary = cue_freq.merge(fail_counts, left_on="cue", right_index=True, how="left").fillna(0)
summary["fail_rate_given_cue_%"] = (summary["fails"] / summary["count"] * 100).round(1)
print("\nCues con mayor tasa de fallo (pistas presentes pero sin extracci√≥n):")
display(summary.sort_values("fail_rate_given_cue_%", ascending=False).head(20))
```
